### Question 3

Write code which will reproduce each of the following patterns given an input size N

Example:

```
N = 10

 x x x x x x x x x x
 x o x o x o x o x o
 x x x x x x x x x x
 x o x o x o x o x o
 x x x x x x x x x x
 x o x o x o x o x o
 x x x x x x x x x x
 x o x o x o x o x o
 x x x x x x x x x x
 x o x o x o x o x o

```

````python
N = 11
pattern = [[' x' for x in range(N)] for y in range(N)]

for j in range(N):
    for i  in range(N):
        if j%2 == 1 or i%2 == 1:
            pattern[j][i] = ' o'

for row in pattern:
    print(''.join(row))
    
# Another approch is to use the list comprehension to define everything (this is not necessarily easier)
print('-'*(N*2))

pattern = [[' o' if y%2 == 1 or x%2 == 1 else ' x' for x in range(N)] for y in range(N)]

for row in pattern:
    print(''.join(row))
````
